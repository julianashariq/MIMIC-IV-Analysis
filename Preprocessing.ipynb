# Data Cleaning (Remove outliers)


*   Normalize the results
*   Draw boxplot to visualize the outliers



```python
#Determine if data is skewed
#A skewness value of 0 in the output denotes a symmetrical distribution of values.
# A negative skewness value indicates an asymmetry in the distribution and the tail is larger towards the left hand side of the distribution.
# A positive skewness value indicates an asymmetry in the distribution and the tail is larger towards the right hand side of the distribution.
clean2_df.skew()
```

```python
clean3_df = clean2_df
X_raw = clean3_df.iloc[:,:23]
y_df = clean3_df.iloc[:,[23]]

for x in list(X_raw):
      if x in ["gender","sofa_cardiovascular", "sofa_cns","sofa_respiration","charlson_comorbidity_index"]:
        continue
      else:
        #q75,q25 = np.percentile(X_raw.loc[:,x],[75,25])
        #intr_qr = q75-q25
        #max = q75+(1.5*intr_qr)
        #min = q25-(1.5*intr_qr)
        q98,q2 = np.percentile(X_raw.loc[:,x],[98,2])
        max = q98
        min = q2
        X_raw.loc[X_raw[x] < min,x] = min
        X_raw.loc[X_raw[x] > max,x] = max

X_raw.describe()
```

```python
# draw box plot to visualize outliers
X_raw.boxplot(fontsize=8,figsize=(20,10))
```

# Feature Scaling
Range of values can influence how each feature is weighted by the ML algorithm. Feature scaling can be performed via:
*   normalization of features (does not assume normal distribution, susceptible to outliers)
*   standardization (assumes normal distribution, less susceptible to outliers)



```python
#normalizing the features
X_df = (X_raw-X_raw.min())/(X_raw.max()-X_raw.min())
X_df.describe().transpose()
```

# Balance Outcome Class
Algorithm will be unable to predict outcome accurately if there is an unbalanced outcome class

```python
# The original outcome statistics.
y_df.value_counts()
```

```python
#Split dataset to training and test dataset. Balancing of outcome should only be done on training dataset
#Split into 6:4 dataset
cases = ['train','test']

#generates a array of length(X_df) with randomly allocated train or test
case_list = np.random.choice(cases,size=len(X_df),replace=True,p=[0.6,0.4])

# returning rows in X_df and y_df that corresponds to the index of the case_list of train and test
X_df_Train, y_df_Train = X_df.iloc[case_list=='train'], y_df.iloc[case_list=='train']
X_df_Test, y_df_Test = X_df.iloc[case_list=='test'], y_df.iloc[case_list=='test']
```

```python
# Balance outcomes using SMOTE
from imblearn.over_sampling import SMOTE
smote_sampler = SMOTE(random_state=12,sampling_strategy='minority')
X_df_SMOTE, y_df_SMOTE = smote_sampler.fit_resample(X_df_Train, y_df_Train)
```
